```python
all_columns_games_df.columns
```

```python
all_columns_games_df = all_columns_games_df.rename(columns = {'GAME_DATE_EST':'GAME_DATE', 'PTS_home': 'POINTS_HOME', 'FG_PCT_home': '2pointer_PCT_HOME', 'FT_PCT_home':'FreeThrow_PCT_HOME',
                                                 'FG3_PCT_home':'3pointer_PCT_HOME', 'AST_home':'Assists_HOME', 
                                                  'REB_home':'Rebounds_HOME', 'FG_PCT_away': '2pointer_PCT_AWAY', 'FT_PCT_away':'FreeThrow_PCT_AWAY',
                                                 'FG3_PCT_away':'3pointer_PCT_AWAY', 'AST_away':'Assists_AWAY', 
                                                  'REB_away':'Rebounds_AWAY', 'PTS_away': 'POINTS_AWAY', 'FG_PCT_away': '2pointer_PCT_AWAY', 'FT_PCT_away':'FreeThrow_PCT_AWAY',
                                                 'FG3_PCT_away':'3pointer_PCT_AWAY', 'AST_away':'Assists_AWAY', 
                                                  'REB_away':'Rebounds_AWAY'})
all_columns_games_df.head()
```

```python
all_columns_games_df.columns
```

```python
home_team_games = all_columns_games_df[['GAME_DATE', 'SEASON', 'HOME_TEAM_ID', 'HOME_TEAM_NAME', 'POINTS_HOME', '2pointer_PCT_HOME',
                                        '3pointer_PCT_HOME', 'FreeThrow_PCT_HOME', 'Assists_HOME', 'Rebounds_HOME', 'HOME_TEAM_WINS']]
home_team_games
```

```python
total_wins_by_home_team_trial = home_team_games.groupby(['HOME_TEAM_NAME', 'SEASON']).agg({'HOME_TEAM_ID': 'last', 'POINTS_HOME': 'sum',
                                                                                     '2pointer_PCT_HOME': 'mean', '3pointer_PCT_HOME': 'mean', 
                                                                                     'FreeThrow_PCT_HOME': 'mean', 'Assists_HOME': 'sum', 'Rebounds_HOME': 'sum',
                                                                                     'HOME_TEAM_WINS': 'sum'}).reset_index()
total_wins_by_home_team_trial
```

```python
total_wins_by_home_team = home_team_games.groupby(['HOME_TEAM_NAME', 'SEASON']).agg({'HOME_TEAM_ID': 'last', 'POINTS_HOME': 'sum',
                                                            'Assists_HOME': 'sum', 'Rebounds_HOME': 'sum',
                                                            'HOME_TEAM_WINS': 'sum'}).reset_index()
total_wins_by_home_team.head(30)
```

```python
away_team_games = all_columns_games_df[['GAME_DATE', 'SEASON', 'VISITOR_TEAM_ID', 'VISITOR_TEAM_NAME', 'POINTS_AWAY', '2pointer_PCT_AWAY',
                                        '3pointer_PCT_AWAY', 'FreeThrow_PCT_AWAY', 'Assists_AWAY', 'Rebounds_AWAY', 'HOME_TEAM_WINS']]
away_team_games
```

```python
total_wins_by_away_team = away_team_games.groupby(['VISITOR_TEAM_NAME', 'SEASON']).agg({'VISITOR_TEAM_ID': 'last', 'POINTS_AWAY': 'sum',
                                                            'Assists_AWAY': 'sum', 'Rebounds_AWAY': 'sum',
                                                            'HOME_TEAM_WINS': 'sum'}).reset_index()
total_wins_by_away_team
```

```python
total_wins_by_away_team.dtypes
```

```python
wins = []
for index,row in total_wins_by_away_team.iterrows():
    if(row["SEASON"] == 2011):
        wins.append(33 - row['HOME_TEAM_WINS'])
    elif(row["SEASON"] == 2020):
        wins.append(36 - row['HOME_TEAM_WINS'])
    else:
        wins.append(41 - row['HOME_TEAM_WINS'])
print(wins)
```

```python
total_wins_by_away_team['AWAY_TEAM_WINS'] = wins
```

```python
total_wins_by_away_team = total_wins_by_away_team.drop(['HOME_TEAM_WINS'], axis = 1)
total_wins_by_away_team
```
