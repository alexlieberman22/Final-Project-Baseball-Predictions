# Draft Data

```python
import sqlite3
import pandas as pd

dat = sqlite3.connect('basketball.sqlite')
query = dat.execute("SELECT * From Draft")
cols = [column[0] for column in query.description]
draft = pd.DataFrame.from_records(data = query.fetchall(), columns = cols)
```

```python
draft
```

```python
print(draft.columns)
```

```python
draft.isnull().sum()
```

```python
draft[draft['nameOrganizationFrom'].isnull()]
```

```python
draft_data = draft[['yearDraft', 'numberPickOverall', 'numberRound',
       'namePlayer', 'slugTeam', 'nameOrganizationFrom',
       'idPlayer', 'idTeam',
       'teamName', 'slugOrganizationTypeFrom',
       'locationOrganizationFrom']]
```

```python
draft_data
```

```python
draft_df = draft_data.rename(columns = {'yearDraft':'DRAFT_YEAR', 'numberPickOverall':'DRAFT_NUMBER', 
                                        'numberRound':'DRAFT_ROUND','namePlayer':'PLAYER_NAME', 
                                        'slugTeam':'TEAM_ABBREVIATION', 'nameOrganizationFrom':'TEAM_NAME',
                                        'idPlayer':'PLAYER_ID', 'idTeam':'TEAM_ID','teamName':'PRO_TEAM_NAME', 
                                        'slugOrganizationTypeFrom':'COLLEGE_OR_PRO','locationOrganizationFrom':'PRO_COUNTRY'})
draft_df
```

```python
len(draft_df)
```

```python
draft_df.isna().sum()
```

```python
draft_df['TEAM_NAME'] = draft_df['TEAM_NAME'].fillna('No Team')
```

```python
draft_df['PRO_COUNTRY'] = draft_df['PRO_COUNTRY'].fillna('Collegiate Player')
```

```python
draft_df
```

```python
draft_df.isnull().sum()
```

```python
draft_df.dtypes
```

```python
draft_df[draft_df['PRO_COUNTRY'] == '']
```

```python
draft_df.to_csv('../Resources/Draft_data.csv')
```
